###  here is the thing, first we make a function that identifies similar ids in a particular eg. deposits. Then it counts how many duplicates are there in the particular column and add it to the newly created dataframe with new column in it "deposits_count" and similarly it goes for amount and convert it into a float value as it may be a string in some rows. and count each and every duplicate rows. like wise calculate for other in the cell. then we move to the following columns

```python
def process_action_column(df, column_name):
    wallet_stats = defaultdict(lambda: {"count": 0, "total_amount": 0.0, "total_usd": 0.0})

    for cell in df[column_name].dropna():
        # Make it iterable (list) if it's just a single dict
        entries = [cell] if isinstance(cell, dict) else cell

        for entry in entries:
            try:
                account_id = entry['account']['id']
                amount = float(entry.get('amount', 0))
                amount_usd = float(entry.get('amountUSD', 0))

                wallet_stats[account_id]["count"] += 1
                wallet_stats[account_id]["total_amount"] += amount
                wallet_stats[account_id]["total_usd"] += amount_usd
            except Exception as e:
                print(f"Skipping entry: {entry} — Error: {e}")

    result_df = pd.DataFrame([
        {
            "account_id": k,
            f"{column_name}_count": v["count"],
            f"{column_name}_total_amount": v["total_amount"],
            f"{column_name}_total_usd": v["total_usd"]
        }
        for k, v in wallet_stats.items()
    ])

    return result_df
```

### The liquidates column's format is little differnt. so we have to make another function which extracts features properly.

```python
def process_liquidates_column(df, column_name):
    features = {}

    for cell in df[column_name].dropna():
        # Make sure cell is a list of dicts
        if isinstance(cell, str):
            try:
                cell = eval(cell)
            except:
                continue

        if isinstance(cell, dict):
            cell = [cell]

        for entry in cell:
            if not isinstance(entry, dict):
                continue

            liquidator = entry.get("liquidator", {})
            account_id = liquidator.get("id")
            if not account_id:
                continue

            try:
                amount = float(entry.get("amount", 0))
                amount_usd = float(entry.get("amountUSD", 0))
            except:
                amount = 0.0
                amount_usd = 0.0

            if account_id not in features:
                features[account_id] = {
                    "liquidates_count": 0,
                    "liquidates_total": 0.0,
                    "liquidates_total_usd": 0.0
                }

            features[account_id]["liquidates_count"] += 1
            features[account_id]["liquidates_total"] += amount
            features[account_id]["liquidates_total_usd"] += amount_usd

    # Turn into DataFrame
    df_features = pd.DataFrame.from_dict(features, orient='index').reset_index()
    df_features = df_features.rename(columns={'index': 'account_id'})

    # Add avg column
    df_features["liquidates_avg_usd"] = df_features["liquidates_total_usd"] / df_features["liquidates_count"]

    return df_features
```

```python
df_liquidates_features = process_liquidates_column(df, "liquidates")
df_liquidates_features.value_counts()
```

# Merging all the seperate dataframe into one single datasframe

```python
from functools import reduce

# List of all the processed feature DataFrames
dfs_to_merge = [
    df_deposits_features,
    df_withdraws_features,
    df_borrows_features,
    df_repays_features,
    df_liquidates_features
]

# Merge them one by one on account_id
df_merged = reduce(lambda left, right: pd.merge(left, right, on="account_id", how="outer"), dfs_to_merge)

# Fill NaNs with 0 (means no activity of that type)
df_merged = df_merged.fillna(0)
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Scale the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_merged.drop("account_id", axis=1))

# Using Elbow method to find best k value (number of clusters)
inertia = []
k_range = range(1, 10)

for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(k_range, inertia, marker='o')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
# Apply KMeans with the chosen k
kmeans = KMeans(n_clusters=4, random_state=42)
df_merged["cluster"] = kmeans.fit_predict(X_scaled)
```

```python
# Pick the metric to rank clusters by
cluster_scores = df_merged.groupby("cluster")["deposits_total_usd"].mean()

# Rank clusters (higher mean USD = better)
cluster_ranks = cluster_scores.rank(method="dense", ascending=False).astype(int)

# Normalize ranks to a 1–100 scale
max_rank = cluster_ranks.max()
score_map = cluster_ranks.apply(lambda r: int((r / max_rank) * 100))

# Map each wallet’s cluster to a score
df_merged["wallet_score"] = df_merged["cluster"].map(score_map)
```

```python
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.histplot(df_merged["wallet_score"], bins=20, kde=True, color="purple")
plt.title("Distribution of Wallet Scores (1–100)")
plt.xlabel("Wallet Score")
plt.ylabel("Count")
plt.show()
```

```python
cluster_summary = df_merged.groupby("cluster").mean(numeric_only=True).sort_values("wallet_score", ascending=False)
cluster_summary[["deposits_count", "withdraws_count", "liquidates_count", "wallet_score"]]
```

```python
def calculate_wallet_score(row):
    # Initialize score based on cluster
    cluster_score_map = {2: 100, 0: 75, 1: 50, 3: 25}
    score = cluster_score_map.get(row['cluster'], 50)  # Default to 50 if no cluster

    # Adjust score based on transaction activity
    if row['deposits_count'] > 50:  # If deposits count is high
        score += 10
    if row['withdraws_count'] > 50:  # If withdrawals count is high
        score += 10

    # Adjust score based on liquidates (penalize if liquidates are high)
    if row['liquidates_count'] > 10:
        score -= 20

    # Normalize the score to be between 1 and 100 (if it goes over 100, we cap it)
    score = max(1, min(score, 100))

    return score

# Apply the function to your dataframe
df_merged['wallet_score'] = df_merged.apply(calculate_wallet_score, axis=1)

# Check the updated dataframe
print(df_merged[['cluster', 'deposits_count', 'withdraws_count', 'liquidates_count', 'wallet_score']].head())
```

```python
# Visualize wallet score distribution
plt.figure(figsize=(10, 6))
plt.hist(df_merged['wallet_score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Wallet Score Distribution')
plt.xlabel('Wallet Score')
plt.ylabel('Frequency')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Drop non-numeric stuff (like wallet address if present)
X = df_merged.drop(columns=["cluster", "wallet_score"], errors='ignore')
X = X.select_dtypes(include='number')  # keep only numeric columns

# PCA to 2D
pca = PCA(n_components=2)
pca_result = pca.fit_transform(X)

# Add PCA result back to df
df_plot = df_merged.copy()
df_plot["pca_1"] = pca_result[:, 0]
df_plot["pca_2"] = pca_result[:, 1]

# Plot
plt.figure(figsize=(10, 6))
sns.scatterplot(
    x="pca_1",
    y="pca_2",
    hue="wallet_score",
    size="liquidates_count",
    data=df_plot,
    palette="viridis",
    alpha=0.8,
    sizes=(20, 200)
)
plt.title("Clustering Interpretation: Wallet Behavior")
plt.legend(bbox_to_anchor=(1.05, 1), loc=2)
plt.tight_layout()
plt.show()
```
